Alzheimer's Disease Detection Using Lightweight CNN on MRI Scans

Data Loading and Exploration

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import confusion_matrix, classification_report
import pyarrow.parquet as pq
from PIL import Image
import io
import base64
```

```python
# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
```

```python
# Display versions
print(f"TensorFlow version: {tf.__version__}")
print(f"Python version: {pd.__version__}")
```

```python
# Define path to your dataset
train_path = 'Data/train-00000-of-00001-c08a401c53fe5312.parquet'
test_path = 'Data/test-00000-of-00001-44110b9df98c5585.parquet'

# Function to load parquet files
def load_parquet_dataset(file_path):
    table = pq.read_table(file_path)
    df = table.to_pandas()
    return df

# Load the datasets
train_df = load_parquet_dataset(train_path)
test_df = load_parquet_dataset(test_path)

# Pre-process DataFrame
train_df['image'] = train_df['image'].apply(lambda d: d['bytes'])
test_df['image'] = test_df['image'].apply(lambda d: d['bytes'])

# Display some information about the datasets
print(f"Training set shape(rows, colums): {train_df.shape}")
print(f"Test set shape(rows, colums): {test_df.shape}")
print("\nTraining set columns:", train_df.columns.tolist())
```

```python
# Convert label ids to class names
class_names = {
    0: 'Mild_Demented',
    1: 'Moderate_Demented',
    2: 'Non_Demented',
    3: 'Very_Mild_Demented'
}

# Check class distribution in training data
train_class_dist = train_df['label'].value_counts().sort_index()
train_class_dist.index = train_class_dist.index.map(lambda x: class_names[x])

# Check class distribution in test data
test_class_dist = test_df['label'].value_counts().sort_index()
test_class_dist.index = test_class_dist.index.map(lambda x: class_names[x])

# Plot class distribution
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.barplot(x=train_class_dist.index, y=train_class_dist.values)
plt.title('Class Distribution - Training Set')
plt.xticks(rotation=45)
plt.ylabel('Number of Samples')

plt.subplot(1, 2, 2)
sns.barplot(x=test_class_dist.index, y=test_class_dist.values)
plt.title('Class Distribution - Test Set')
plt.xticks(rotation=45)
plt.ylabel('Number of Samples')

plt.tight_layout()
plt.show()
```

```python
# Shows 4 sample images from the training set, one row per dementia class.
# Function to convert binary image data to numpy array
# def decode_image(image_bytes):
#     img = Image.open(io.BytesIO(image_bytes))
#     return np.array(img)

def decode_image(image_bytes):
    # Decode if the image is base64 encoded (optional, depending on your dataset)
    # image_bytes = base64.b64decode(image_bytes)  # Uncomment if needed

    img = tf.io.decode_jpeg(image_bytes, channels=3)  # Ensures 3 channels
    return img

# Display some sample images
def display_samples(df, num_samples=4):
    plt.figure(figsize=(15, 3*num_samples))
    samples_per_class = {}
    
    # Track how many samples we've shown for each class
    for i, (idx, row) in enumerate(df.iterrows()):
        class_id = row['label']
        class_name = class_names[class_id]
        
        # Initialize counter for this class if not already done
        if class_name not in samples_per_class:
            samples_per_class[class_name] = 0
            
        # Skip if we already have enough samples for this class
        if samples_per_class[class_name] >= num_samples:
            continue
            
        # Increment counter
        samples_per_class[class_name] += 1

        # Plot the image
        plt.subplot(4, num_samples, samples_per_class[class_name] + (list(class_names.values()).index(class_name) * num_samples))
        img = decode_image(row['image'])
        plt.imshow(img)
        plt.title(f"{class_name}")
        plt.axis('off')
        
        # Break if we have enough samples for each class
        if all(samples_per_class.get(c, 0) >= num_samples for c in class_names.values()):
            break
            
    plt.tight_layout()
    plt.show()

# Display sample images from the training set
display_samples(train_df, num_samples=4)
```

Data Preprocessing and Augmentation

```python
# Define image dimensions (as mentioned in the paper)
IMG_HEIGHT = 150
IMG_WIDTH = 150
BATCH_SIZE = 50  # From the paper

# Function to preprocess and create datasets
def prepare_dataset(df, is_training=True):
    images = []
    labels = []
    
    for _, row in df.iterrows():
        # Decode and preprocess image
        img = decode_image(row['image'])

        # Debug: Ensure the image is 3D before resizing
        assert len(img.shape) == 3, f"Image shape was {img.shape}, expected 3D"
        img = tf.image.resize(img, [IMG_HEIGHT, IMG_WIDTH])
        img = img / 255.0  # Normalize pixel values
        
        images.append(img)
        labels.append(row['label'])
    
    images = np.array(images)
    labels = np.array(labels)
    
    return images, labels

# Prepare training and testing datasets
X_train, y_train = prepare_dataset(train_df, is_training=True)
X_test, y_test = prepare_dataset(test_df, is_training=False)

print(f"Training data shape: {X_train.shape}")
print(f"Training labels shape: {y_train.shape}")
print(f"Testing data shape: {X_test.shape}")
print(f"Testing labels shape: {y_test.shape}")
```

```python
# Data augmentation setup (as described in the paper)
datagen = ImageDataGenerator(
    rescale=1./255,
    brightness_range=[0.8, 1.2],
    zoom_range=0.1,
    fill_mode='constant',
    horizontal_flip=True
)

# Handle class imbalance by generating more samples for underrepresented classes
def augment_minority_classes(X, y, target_counts=None):
    """
    Augment minority classes to achieve balanced classes
    
    Parameters:
    - X: Images
    - y: Labels
    - target_counts: Dictionary mapping class ids to target counts
    
    Returns:
    - Augmented images and labels
    """
    # Find the class with the most samples if target_counts not provided
    if target_counts is None:
        class_counts = np.bincount(y)
        max_count = np.max(class_counts)
        target_counts = {i: max_count for i in range(len(class_counts))}
    
    # Create augmented datasets
    X_augmented = []
    y_augmented = []
    
    for class_id, target_count in target_counts.items():
        # Get samples of this class
        class_indices = np.where(y == class_id)[0]
        class_samples = X[class_indices]
        current_count = len(class_indices)
        
        # Add original samples
        X_augmented.append(class_samples)
        y_augmented.append(np.full(current_count, class_id))
        
        # Add augmented samples if needed
        if current_count < target_count:
            # How many augmented samples we need
            augment_count = target_count - current_count
            
            # Generate augmented images
            auggen = ImageDataGenerator(
                rotation_range=15,
                width_shift_range=0.1,
                height_shift_range=0.1,
                zoom_range=0.1,
                horizontal_flip=True,
                fill_mode='nearest'
            )
            
            # Reshape to match expected input
            reshaped_samples = class_samples.reshape(current_count, IMG_HEIGHT, IMG_WIDTH, 3)
            
            # Generate augmented samples
            aug_samples = []
            aug_iter = auggen.flow(reshaped_samples, batch_size=1)
            
            for _ in range(augment_count):
                # aug_samples.append(aug_iter.next()[0])
                aug_samples.append(next(aug_iter)[0])
            
            # Add to augmented dataset
            X_augmented.append(np.array(aug_samples))
            y_augmented.append(np.full(augment_count, class_id))
    
    # Combine all samples
    X_balanced = np.vstack(X_augmented)
    y_balanced = np.concatenate(y_augmented)
    
    # Shuffle the augmented dataset
    indices = np.arange(len(y_balanced))
    np.random.shuffle(indices)
    X_balanced = X_balanced[indices]
    y_balanced = y_balanced[indices]
    
    return X_balanced, y_balanced

# Check class distribution
class_counts = np.bincount(y_train)
print("Class distribution before augmentation:", class_counts)

# Target count for each class (using maximum count as reference)
max_count = np.max(class_counts)
target_counts = {i: max_count for i in range(len(class_counts))}

# Apply augmentation only for training data
X_train_augmented, y_train_augmented = augment_minority_classes(X_train, y_train, target_counts)

print("Class distribution after augmentation:", np.bincount(y_train_augmented))
```
